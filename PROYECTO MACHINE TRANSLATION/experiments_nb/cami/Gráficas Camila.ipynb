```python
def graphIE(historiales, to_graph):
    h=1
    for historial in historiales:
        v=0
        
        for variante in historial:
            v=v+1
            for x in [to_graph]:
                
                plt.plot(variante[x],label=f"Historic {h} vers {v}")
        h=h+1
    plt.legend()
    plt.xlabel("Epoch",fontsize=13)
    plt.ylabel(to_graph,fontsize=13)
    
```
